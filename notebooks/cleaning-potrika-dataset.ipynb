```python
import re
```

```python
def alpha_num_bengali(text):
    return re.sub(r'[^A-Za-z0-9\u0980-\u09FF\u09E6-\u09EF ]', '', text)
```

```python
def remove_bengali_numerics(text):
    return re.sub(r'[\u09E6-\u09EF]', '', text)
```

```python
def remove_brackets(text):
    return re.sub(r'\(\s*[^()]*\)', '', text)
```

```python
def remove_links(text):
    return re.sub(r'https?:\/\/\S+|www\.\S+', '', text)
```

```python
def remove_newLine(text):
    return re.sub(r'[\r\n]+', ' ', text)
```

```python
def remove_engNum(text):
    return re.sub(r'\d+', '', text)
```

### now let's remove stopwords

```python
def get_column_values(file_path, sheet_name, column_name):
    # Read the Excel file
    df = pd.read_excel(file_path, sheet_name=sheet_name)
    
    # Extract the specific column and convert it to a list
    column_values = df[column_name].dropna().tolist()
    
    return column_values

# Example usage
file_path = 'potrika/stopwords/stopwords_bangla.xlsx'  # Replace with the path to your .xlsx file
sheet_name = 'Sheet1'       # Replace with the sheet name you want to read
column_name = 'words'     # Replace with the name of the column you want to extract

# Get the list of values from the specified column
bn_sw = get_column_values(file_path, sheet_name, column_name)
print(bn_sw)
```

```python
def remove_stopwords(text):
    final_text = []
    for i in text.split():
        if i.strip() not in bn_sw:
            final_text.append(i.strip())
    return " ".join(final_text)
```

```python
data['article'] = data['article'].apply(remove_stopwords)
```
